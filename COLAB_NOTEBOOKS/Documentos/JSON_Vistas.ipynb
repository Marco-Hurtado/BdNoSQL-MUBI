```python
vistas = pd.read_csv(ruta_csv+'Vistas.csv')
usuarios = pd.read_csv(ruta_csv+'Usuarios.csv')
peliculas = pd.read_csv(ruta_csv+'Peliculas.csv')
pelicula_director = pd.read_csv(ruta_csv+'Pelicula_director.csv')
pelicula_genero = pd.read_csv(ruta_csv+'Pelicula_genero.csv')
pelicula_actor = pd.read_csv(ruta_csv+'Pelicula_actor.csv')
productoras = pd.read_csv(ruta_csv+'Productoras.csv')
directores = pd.read_csv(ruta_csv+'Directores.csv')
generos = pd.read_csv(ruta_csv+'Generos.csv')
actores = pd.read_csv(ruta_csv+'Actores.csv')
paises = pd.read_csv(ruta_csv+'Paises.csv')
```

```python
mamastrofico = vistas.merge(usuarios, how='left', on='id_usuario')
```

```python
mamastrofico = mamastrofico.merge(peliculas, how='left', on='id_pelicula')
```

```python
mamastrofico = mamastrofico.merge(paises, how='left', on='id_pais')
mamastrofico = mamastrofico.merge(productoras, how='left', on='id_productora', suffixes=['_pais','_productora'])
```

Lista de actores

```python
lista_actores = pelicula_actor.merge(actores, how='left', on='id_actor')
```

```python
lista_actores = lista_actores.drop('id_actor', axis=1)
```

```python
lista_actores = lista_actores.groupby('id_pelicula').agg(lambda x: list(x))
```

Lista de directores

```python
lista_directores = pelicula_director.merge(directores, how='left', on='id_director')
lista_directores = lista_directores.drop('id_director', axis=1)
lista_directores = lista_directores.groupby('id_pelicula').agg(lambda x: list(x))
```

Lista de generos

```python
lista_generos = pelicula_genero.merge(generos, how='left', on='id_genero')
lista_generos = lista_generos.drop('id_genero', axis=1)
lista_generos = lista_generos.groupby('id_pelicula').agg(lambda x: list(x))
```

Unir las listas al mamastrofico

```python
mamastrofico = mamastrofico.merge(lista_actores, how='left', on='id_pelicula')
mamastrofico = mamastrofico.merge(lista_directores, how='left', on='id_pelicula', suffixes=['_actores', '_directores'])
```

```python
mamastrofico = mamastrofico.merge(lista_generos, how='left', on='id_pelicula')
```

```python
mamastrofico
```

```python
mamastrofico.columns = [x.replace('nombre_', '') for x in mamastrofico.columns]
mamastrofico = mamastrofico.rename({'nombre': 'generos'}, axis=1)
mamastrofico = mamastrofico.drop([x for x in mamastrofico.columns if x.startswith('id_')], axis=1)
```

```python
mamastrofico.columns
```

```python
aggs_list = {'fecha': list,
             'email': list,
             'usuario': list,
             'genero': list,
             'fecha_nacimiento': list,
             'anio': lambda x: list(x)[0],
             'duracion': lambda x: list(x)[0],
             'descripcion': lambda x: list(x)[0],
             'actores': lambda x: list(x)[0],
             'directores': lambda x: list(x)[0],
             'generos': lambda x: list(x)[0]}
```

```python
grouped = mamastrofico.groupby(['pais', 'productora', 'titulo']).agg(aggs_list)
```

```python
grouped
```

```python
i = 0
total = grouped.shape[0]

lista_paises = []
json_dict = {'paises': []}

pais_ant, productora_ant, titulo_ant = '','',''

for x in grouped.iterrows():

  if i % 180==0: print(f'{100*i/total}% completado')
  i+=1

  pais, productora, titulo = x[0]
  if pais != pais_ant:
    json_dict['paises'].append({'pais': pais, 'productoras': []})
    json_dict['paises'][-1]['productoras'].append({'productora': productora, 'peliculas': []})
  elif productora != productora_ant:
    json_dict['paises'][-1]['productoras'].append({'productora': productora, 'peliculas': []})

    # json_dict[pais]['productoras'].append({'nombre': productora, 'peliculas': []})

  lista_vistas = [{'usuario': {'email': e, 'nombre': n, 'fecha_nacimiento': f_n, 'genero': g},
                     'fecha': f} for e,n,f_n,g,f in zip(x[1]['email'],
                                                                x[1]['usuario'],
                                                                x[1]['fecha_nacimiento'],
                                                                x[1]['genero'],
                                                                x[1]['fecha'])]

  pelicula_info = {'titulo': titulo,
                   'generos': x[1]['generos'],
                   'directores': x[1]['directores'],
                   'actores': x[1]['actores'],
                   'vistas':lista_vistas}


  json_dict['paises'][-1]['productoras'][-1]['peliculas'].append(pelicula_info)

  pais_ant, productora_ant, titulo_ant = pais, productora, titulo
```

```python
with open(ruta_json+'vistas.json', 'w') as f:
  json.dump(json_dict, f, indent=4)
```
