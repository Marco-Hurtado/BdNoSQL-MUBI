```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
peliculas = pd.read_csv('/content/drive/MyDrive/CDD/7mo/Bases/MUBI/CSV_FILES/Relacional/Peliculas.csv')
```

```python
usuarios = pd.read_csv('/content/drive/MyDrive/CDD/7mo/Bases/MUBI/CSV_FILES/Relacional/Usuarios.csv')
```

```python
catalogo_peliculas = peliculas['id_pelicula']
catalogo_usuarios = usuarios['id_usuario']
```

```python
vistas = pd.DataFrame(columns = ['id_vista', 'id_usuario', 'id_pelicula', 'fecha'])
```

```python
tamanio = 7_500_000
```

```python
vistas['id_vista'] = np.arange(tamanio)
```

```python
vistas['id_usuario'] = np.random.choice(catalogo_usuarios, tamanio)
```

```python
n = catalogo_peliculas.shape[0]
n
```

```python
index = np.arange(n)
mean = n / 2
std_dev = n / 1000
probabilities = np.exp(-0.5 * ((index - mean) / std_dev)**2)

# Normalize the probabilities
probabilities /= probabilities.sum()
```

```python
vistas['id_pelicula'] = np.random.choice(catalogo_peliculas, tamanio, p=probabilities)
```

```python
start = np.datetime64('2008-02-01')
end = np.datetime64('2020-04-30')

# Generate random dates
vistas['fecha'] = np.random.uniform(start.astype('datetime64[D]').astype(int),
                                  end.astype('datetime64[D]').astype(int),
                                  tamanio).astype('datetime64[D]')
```

```python
vistas = vistas.sort_values(by='fecha')
vistas
```

```python
vistas.info(verbose=True)
```

```python
vistas.to_csv('/content/drive/MyDrive/CDD/7mo/Bases/MUBI/CSV_FILES/Relacional/Vistas.csv', index=False)
```
